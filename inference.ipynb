```python
import pandas as pd
path_to_test_data = "../../data/"
output_directory = "fasta_files"
test_data = pd.read_csv(path_to_test_data+"test_sequences.csv")

def generate_fasta_file(test_data):
    sequences = test_data["sequence"].tolist()
    target_id = test_data["target_id"].tolist()
    for seq, target in zip(sequences, target_id):
        with open(f"{output_directory}/{target}.fasta", "a") as f:
            f.write(f">{target}\n{seq}\n")
            
generate_fasta_file(test_data)
```

## We will run just one iteration of MSA2xyz to understand what are the inputs/outputs.

### MSA2xyz iteration generates a Pair Matrix and MSA features and those are given to the Structure Module

```python
from cfg_97 import EvoMSA2XYZ
from cfg_97.EvoMSA2XYZ import MSA2xyzIteration

# one-hot encoded dims - [A, U, G, C, gap (-), mask, padding]
msa_dim= 7
# one-hot encoded dims - [A, U, G, C, gap (-), mask]
seq_dim = 6
#embeddings dims
m_dim,s_dim,z_dim = 64,64,64
#N ensembles and N cycles
N_ensemble,N_cycle=1,1
#Length of the sequence
L = msa.size(1)

#Previous distance: pairwise distances between nucleotides - OUTPUT of the structure module (from one cycle)
## [0-2Å], [2-2.5Å], [2.5-3Å], ..., [>20Å]
previous_dis = torch.zeros([L,L,38]).to(msa.device) 
previous_dis[...,0] = 1

#Previous helix-bond - OUTPUT of the structure module (from one cycle)
#Types of helix bonds: A-U pair, G-C pair, Wobble G-U, Non-canonical, Stacking, no interaction?
previous_hb = torch.zeros([L,L,6]).to(msa.device)
previous_hb[...,0] = 1

m1_pre,z_pre=0,0
x_pre=torch.zeros(L,3,3).to(msa.device)
index = 0 

model = MSA2xyzIteration(seq_dim,msa_dim,N_ensemble,m_dim,s_dim,z_dim)
model = model.to(DEVICE)
model.eval()

# Move model to the specified device
alphas = np.array(list(seq))
previous_dis = previous_dis.to(DEVICE) 
previous_hb = previous_hb.to(DEVICE)

# Now make prediction
preds = model.pred(msa, seq_idx, None, m1_pre, z_pre, x_pre, index, alphas, previous_dis, previous_hb)
    
```

```python
m, z, s = preds
print(m.shape, z.shape, s.shape)
```
